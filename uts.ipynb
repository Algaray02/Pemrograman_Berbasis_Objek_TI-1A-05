```python
class AlatElektronik:
   def __init__(self, merk, model):
      self.merk = merk
      self.model = model
      self.status_daya = False
   
   def nyalakan(self):
      self.status_daya = True
      print('Alat Elektronik berhasil dinyalakan...')
   
   def matikan(self):
      self.status_daya = False
      print('Alat Elektronik berhasil dimatikan...')

   def get_info(self):
      print(f'Merk: {self.merk}')
      print(f'Model: {self.model}')
      print(f'Status Daya: {self.status_daya}')

class Televisi(AlatElektronik):
   def __init__(self, merk, model, ukuran_layar):
      super().__init__(merk, model, status_daya)
      self.ukuran_layar = ukuran_layar
   
   def nyalakan(self):
      self.status_daya = True
      print('Televisi berhasil dinyalakan...')
   
   def matikan(self):
      self.status_daya = False
      print('Televisi berhasil dimatikan...')
   
   def get_info(self):
      super().get_info()
      print(f'Ukuran Layar: {self.ukuran_layar} inch')

   def ganti_channel(self, channel):
      if not self.status_daya:
         print('Televisi dalam kondisi mati')
      else:
         print(f'Channel berhasil diganti ke Channel {channel}')

class Radio(AlatElektronik):
   def __init__(self, merk, model, frekuensi):
      super().__init__(merk, model, status_daya)
      self.frekuensi = frekuensi

   def nyalakan(self):
      self.status_daya = True
      print('Radio berhasil dinyalakan...')
   
   def matikan(self):
      self.status_daya = False
      print('Radio berhasil dimatikan...')

   def set_frekuensi(self, frekuensi):
      if not self.status_daya:
         print(f'Radio dalam kondisi mati')
      else:
         print(f'Frekuensi radio berhasil diganti ke {frekuensi} MHz'); self.frekuensi = frekuensi
   
   def get_info(self):
      super().get_info()
      print(f'Frekuensi: {self.frekuensi}')


def main():
   televisi1 = Televisi('Panasonic', 'MT-011', 15)
   radio1 = Radio('Xiaomi', 'RD-001', 5.5)

   list_elektronik = [televisi1, radio1]

   for i in list_elektronik:
      i.get_info()
      i.nyalakan()
      if isinstance(i, Televisi):
         i.ganti_channel(3)
      elif isinstance(i, Radio):
         i.set_frekuensi(9.7)
      else:
         print('Alat elektronik tidak termasuk Televisi maupun Radio...')
      i.matikan()
      print()

if __name__ == "__main__":
   main()


```

```python
class AlatElektronik:
   def __init__(self, merk, model):
      self.merk = merk
      self.model = model
      self.status_daya = False
   
   def nyalakan(self):
      self.status_daya = True
      print(f'{self.merk} berhasil dinyalakan...')
   
   def matikan(self):
      self.status_daya = False
      print(f'{self.merk} berhasil dimatikan...')

   def get_info(self):
      print(f'Merk: {self.merk}')
      print(f'Model: {self.model}')
      print(f'Status Daya: {"Nyala" if self.status_daya else "Mati"}')

class Televisi(AlatElektronik):
   def __init__(self, merk, model, ukuran_layar):
      super().__init__(merk, model)
      self.ukuran_layar = ukuran_layar
   
   def get_info(self):
      super().get_info()
      print(f'Ukuran Layar: {self.ukuran_layar} inch')

   def ganti_channel(self, channel):
      if not self.status_daya:
         print('Televisi dalam kondisi mati')
      else:
         print(f'Channel berhasil diganti ke Channel {channel}')

class Radio(AlatElektronik):
   def __init__(self, merk, model, frekuensi):
      super().__init__(merk, model)
      self.frekuensi = frekuensi

   def set_frekuensi(self, frekuensi):
      if not self.status_daya:
         print(f'Radio dalam kondisi mati')
      else:
         print(f'Frekuensi radio berhasil diganti ke {frekuensi} MHz'); self.frekuensi = frekuensi
   
   def get_info(self):
      super().get_info()
      print(f'Frekuensi: {self.frekuensi}')


def main():
   televisi1 = Televisi('Panasonic', 'MT-011', 15)
   radio1 = Radio('Xiaomi', 'RD-001', 5.5)

   list_elektronik = [televisi1, radio1]

   for i in list_elektronik:
      i.get_info()
      i.nyalakan()
      if isinstance(i, Televisi):
         i.ganti_channel(3)
      elif isinstance(i, Radio):
         i.set_frekuensi(9.7)
      else:
         print('Alat elektronik tidak termasuk Televisi maupun Radio...')
      i.matikan()
      print()

if __name__ == "__main__":
   main()


```
